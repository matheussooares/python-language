# Matplotlib

Partes de uma figura

```python
import matplotlib.pyplot as plt
import numpy as np

n_samples = 100
x = np.arange(0,n_samples+1) 
y_1 = 4 + 3 * x
y_2 = -1/2*x**2 + x + 20 
```

## Plotando

O subplot retorna uma figura e os seus eixos.

```python
fig, ax = plt.subplots(figsize=(5,2.7))

ax.plot(x,y_2,color = 'blue',
        linewidth = 1,
        linestyle = '--')
```
